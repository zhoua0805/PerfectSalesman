```python
from google.colab import files

dataset = files.upload()
```

```python
import io
import pandas as pd

df = pd.read_csv(io.BytesIO(dataset['january.csv']))

df.head()

#len(df[df['Quantity'] > 0 and df['Discount'] > 0])
```

```python
#df = df[df['Quantity'] > 0]

len(df)
```

```python
onehotdf = pd.get_dummies(df['Sub-Category'])
X = pd.concat([df['Discount'], onehotdf], axis=1)
y = df[['Sales', 'Quantity', 'Profit']]

X.head()
```

```python
y.head()
```

```python
import xgboost as xgb
from sklearn.multioutput import MultiOutputRegressor
import numpy as np

np.random.seed(100)

model = MultiOutputRegressor(xgb.XGBRegressor(objective='reg:squarederror', learning_rate=0.001, n_estimators=10)).fit(X.to_numpy(), y.to_numpy())
```

```python
print(np.mean((model.predict(X.to_numpy()) - y.to_numpy())**2, axis=0))
```
